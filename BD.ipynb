# [dbo].[SetCliente]

```sql
/****** Object:  StoredProcedure [dbo].[SetCliente]    Script Date: 11/03/2022 10:10:40 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[SetCliente] 
@pNome varchar(60),
@pSexo char(1),
@pDataNascimento Date,
@pEstadoCivil char(1),
@pCPF char(14),
@pRG char(20),
@Retorno int output
AS  
BEGIN  
   if exists( select * from Cliente where CPF = @pCPF)
   begin
	   select @Retorno = 1;
	   RETURN (@Retorno)
   end
   if exists( select * from Cliente where RG = @pRG)
   begin
	   select @Retorno = 2;
	   RETURN (@Retorno)
   end
   insert into Cliente
 (Nome, Sexo, DataNascimento, EstadoCivil, CPF, RG)
 values
 ( @pNome, @pSexo, @pDataNascimento, @pEstadoCivil, @pCPF, @pRG)
 select @Retorno = 3;
RETURN (@Retorno)
END
GO
```
